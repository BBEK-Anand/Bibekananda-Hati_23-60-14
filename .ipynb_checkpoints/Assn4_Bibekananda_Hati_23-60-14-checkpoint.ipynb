```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('housingdata.csv')
#Q1
df.head(6)
```

```python
#Q2
# print(df.dtypes)
# print(df.index)
df1=pd.DataFrame(df.columns,columns=["Columns"]).T
df2=pd.DataFrame(df.index,columns=["row indeces"])
print(df1,"\n\n",df2)
```

```python
#Q3
print("a)\nTotalno. of \"STATE\" exists in ousingdata.csv file ",df["STATE"].notnull().sum())
print("b)\nNo. of unique states exist in the dataset is",df["STATE"].nunique())
```

```python
#Q4
blank=df.isin({'--',np.nan,'na',"HURLEY"})
blank.values.tolist()
```

```python
#Q5
df=df.dropna(subset=["NUM_BEDROOMS"],inplace=False)
df = df[df["NUM_BEDROOMS"] !='na']
df
```

```python
#Q6
df["NUM_BATH"] =df["NUM_BATH"].replace(to_replace='HURLEY', method='ffill')
df["NUM_BATH"] =df["NUM_BATH"].replace(to_replace=np.nan, method='ffill')
df
```

```python
#Q7
temp=df[df["SQ_FT"] !='--']
mn=temp["SQ_FT"].astype('float32').mean()
df["SQ_FT"]=df['SQ_FT'].replace('--',mn)
df["SQ_FT"]=df['SQ_FT'].replace(np.nan,mn)
df
```

```python
#Q8
df=df.astype({"PRICE":'float32',"NUM_BEDROOMS":'float32',"NUM_BATH":'float32',"SQ_FT":'float32'})
df["SQ_FT"].sum(axis=0)
```

```python
#Q9
df["NUM_STORE"]=1
df
```

```python
#Q14
df[df["SQ_FT"] > 1000]
```

```python
#Q15
df.iloc[:3,:3]
```

```python
#Q16
pbs=pd.DataFrame()
pbs["STATE"]=df["STATE"].values.tolist()
pbs["Price_per_sqft"]=list(df["PRICE"]/df["SQ_FT"])
pbs=pbs.groupby("STATE").mean().sort_values(by="Price_per_sqft",ascending=False)
print("The state having highest average price per square foot of housing area is",pbs.first_valid_index(),".")
```
